## **Expected Monetary Value**

The Expected Monetary Value (EMV) is a key risk management technique used to quantify the potential financial impact of a risk event. It is calculated by multiplying the `probability` of a risk occurring by the `fix_cost` if it does occur. The result is a monetary value that helps prioritize risks and allocate resources for mitigation.


```python
TOTAL_PROJECT_BUDGET = 500000000
for project_type, group_df in df.groupby("Project"):
    group_df["Probability_decimal"] = group_df["Probability"] / 100
    group_df["Fix Cost ($)"] = (group_df["Fix_Cost"] / 100) * TOTAL_PROJECT_BUDGET
    group_df["EMV ($)"] = group_df["Probability_decimal"] * group_df["Fix Cost ($)"]
    total_emv = group_df["EMV ($)"].sum()
    print("-" * 50)
    print(f"Project Type: {project_type}")
    print(f"Total EMV Risk Exposure: ${total_emv:,.2f}")
```

```python
TOTAL_PROJECT_BUDGET = 500000000
results = []
for project_type, group_df in df.groupby("Risk"):
    mask = group_df["Project"] == "Transaction Processing System"
    group_df.loc[mask, "Probability_decimal"] = group_df.loc[mask, "Probability"] / 100
    group_df.loc[mask, "Fix Cost ($)"] = (
        group_df.loc[mask, "Fix_Cost"] / 100
    ) * TOTAL_PROJECT_BUDGET
    group_df.loc[mask, "EMV ($)"] = (
        group_df.loc[mask, "Probability_decimal"] * group_df.loc[mask, "Fix Cost ($)"]
    )
    total_emv = group_df.loc[mask, "EMV ($)"].sum()
    results.append(
        {"Risk Type": project_type, "Total EMV Risk Exposure": f"${total_emv:,.2f}"}
    )
emv_table = pd.DataFrame(results)
display(emv_table)
```
